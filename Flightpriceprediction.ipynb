## Column = 'Date_of_Journey'

```python
# Since the dtype of the column 'Date_of_Journey' is Object let us split the date to avoid problems while performing ML steps
flight_train_test['Date'] = flight_train_test['Date_of_Journey'].str.split('/').str[0]
flight_train_test['Month'] = flight_train_test['Date_of_Journey'].str.split('/').str[1]
flight_train_test['Year'] = flight_train_test['Date_of_Journey'].str.split('/').str[2]
```

# Column = 'Arrival_Time'

```python
# Similarly lets split the 'Arrival_Time' as well and keep the time details
flight_train_test['Arrival_hour'] = flight_train_test['Arrival_Time'].str.split(':').str[0]
flight_train_test['Arrival_minute'] = flight_train_test['Arrival_Time'].str.split(' ').str[1]
```

```python
flight_train_test['Arrival_minute'] = flight_train_test['Arrival_minute'].fillna('00')
```

# Column = 'Dep_Time'

```python
flight_train_test['Dep_hour'] = flight_train_test['Dep_Time'].str.split(':').str[0]
flight_train_test['Dep_min'] = flight_train_test['Dep_Time'].str.split(':').str[1]

flight_train_test['Dep_hour'] = flight_train_test['Dep_hour'].astype(int)
flight_train_test['Dep_min'] = flight_train_test['Dep_min'].astype(int)
```

# Column = 'Total_Stops'

```python
# Checking the values in the column 'Total_Stops'
flight_train_test['Total_Stops'].unique()
```

```python
# Replacing the values in the column 'Total_Stops'
```

```python
flight_train_test['Total_Stops'] = flight_train_test['Total_Stops'].replace('non-stop', 0)
```

```python
flight_train_test['Total_Stops'] = flight_train_test['Total_Stops'].replace('1 stop', 1)
```

```python
flight_train_test['Total_Stops'] = flight_train_test['Total_Stops'].replace('2 stops', 2)
```

```python
flight_train_test['Total_Stops'] = flight_train_test['Total_Stops'].replace('3 stops', 3)
```

```python
flight_train_test['Total_Stops'] = flight_train_test['Total_Stops'].replace('4 stops', 4)
```

```python
# Checking for any abnormal values
flight_train_test['Total_Stops'].unique()
```

```python
# Finding the median of the values in the column
flight_train_test['Total_Stops'].median()
```

```python
# Replacing the nan values with the median
flight_train_test['Total_Stops'] = flight_train_test['Total_Stops'].fillna(1)
```

```python
# Finally checking for the values in the column
flight_train_test['Total_Stops'].unique()
```

```python
# Checking the data type of the column
flight_train_test['Total_Stops'].dtypes
```
